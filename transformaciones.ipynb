```python
import pandas as pd
import numpy as np
```

```python
adidas = pd.read_csv('adidas.csv')
adidas.columns
cols = ['url', 'name','sku', 
        'currency', 'availability','color','category','source',
        'source_website','breadcrumbs','description',
        'brand','images','country','language'
       ]
```

```python
cols_validas = []
cols_many_levels = []
cols_one_level = []


umbral = .98
size_df = adidas.shape[0]

for c_ in cols:
    nivels = adidas.loc[:,c_].drop_duplicates().shape[0]
    
    if (nivels/size_df)>=umbral:
        cols_many_levels.append(c_)
    elif nivels == 1:
        cols_one_level.append(c_)
    else:
        cols_validas.append(c_)
    
```

```python
levels = adidas.groupby( ["color"],as_index=False).agg({"sku": ["count"] })
levels.columns = levels.columns.droplevel(1)
levels
```

```python
levels["Porcentage"] = levels["sku"]/adidas.shape[0]
levels.sort_values(["Porcentage"], ascending=False)
```

```python
levels["color_t"] = "Others"
levels.apply(lambda r: r if r['Porcentage']>0.9 else r['Porcentage'], axis=1)
```

```python
adidas = adidas.merge(levels, on=["color"], how="inner")
```

```python
levels.to_csv("color_categories.csv", index_label=False)
```
